```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df  = pd.read_csv("https://raw.githubusercontent.com/syedayazsa/data/master/food_items.csv",index_col='index')
```

```python
df = df[['brand', 'categories', 'ingredients', 'manufacturer', 'title']]
```

```python
df['brand'] = df['brand'].fillna("")
df['ingredients'] = df['ingredients'].fillna("")
df['manufacturer'] = df['manufacturer'].fillna("")
df['title']= df['title'].fillna("")
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df['brand'].value_counts()
```

```python
df['categories'].value_counts()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df['categories'])
count_matrix.toarray()
similarity_scores = cosine_similarity(count_matrix)
similarity_scores
```

```python
def combine_features(row):
    return row['brand']+" "+row['categories']+" "+row['ingredients']+" "+row['manufacturer']+" "+row['title']
```

```python
df["combined_features"] = df.apply(combine_features,axis =1)
df["combined_features"].head()
```

```python
cv1 = CountVectorizer()
c_matrix = cv1.fit_transform(df["combined_features"])
cos_sim = cosine_similarity(c_matrix)
cos_sim
```
